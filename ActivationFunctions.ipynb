```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def d_sigmoid(x):
    return sigmoid(x)*(1 - sigmoid(x))
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
def d_tanh(x):
    return 1 - tanh(x)**2
```

```python
def relu(x):
    return [max(0, x_val) for x_val in x]
```

```python
def d_relu(x):
    results = []
    for v in x:
        if v >= 0:
            results.append(1)
        else:
            results.append(0)
    return results
```

```python
def leaky_relu(x):
    results = []
    for v in x:
        if v >= 0:
            results.append(v)
        else:
            results.append(0.1*v)
    return results
```

```python
def softmax(x):
    exp_x = np.exp(x)
    return exp_x/sum(exp_x)
```

```python
def e_power_x(x):
    return np.exp(x)
```

```python
def d_softmax(x, dx=0.000001):
    s = softmax(x)
    n = len(x)
    y = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            if i == j:
                y[i, j] = s[i] * (1 - s[j])
            else:
                y[i, j] = -s[i] * s[j]
                
    return y
```

```python
def cross_entropy(x):
    return -np.log(x)
```

```python
def sse(y, y_hat):
    return (y - y_hat)**2
```

```python
def visualization(ax, y_label, x, y):
    ax.plot(x, y)
    ax.set_xlabel('x')
    ax.set_ylabel(y_label)
    ax.grid()
```

```python
def comparation(ax, x_1, y_1, label_1, x_2, y_2, label_2):
    ax.plot(x_1, y_1, label=label_1)
    ax.plot(x_2, y_2, label=label_2)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid()
    ax.legend()
```

```python
def d_relu_visualization(ax):    
    ax.plot(np.linspace(-10., -1, 10), np.linspace(0., 0., 10))
    ax.plot(np.linspace(1., 10, 10), np.linspace(1., 1., 10))
    
    ax.set_xlabel('x')
    ax.set_ylabel('d_ReLU(x)')
    ax.grid()
```

```python
def d_leaky_relu_visualization(ax):    
    ax.plot(np.linspace(-10., -1., 10), np.linspace(0.1, 0.1, 10))
    ax.plot(np.linspace(1., 10, 10), np.linspace(1., 1., 10))
    
    ax.set_xlabel('x')
    ax.set_ylabel('d_ReLU(x)')
    ax.grid()
```

```python
x = np.linspace(-10., 10., 100)
y = sigmoid(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'sigmoid(x)', x, y)
```

```python
x = np.linspace(-10., 10., 100)
y = d_sigmoid(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'd_sigmoid(x)', x, y)
```

```python
x = np.linspace(-10., 10., 100)
y = tanh(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'tanh(x)', x, y)
```

```python
x = np.linspace(-10., 10., 100)
y = d_tanh(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'd_tanh(x)', x, y)
```

```python
x_1 = np.linspace(-10., 10., 100)
y_1 = d_sigmoid(x)

x_2 = np.linspace(-10., 10., 100)
y_2 = d_tanh(x)

fig, ax = plt.subplots(figsize=(4, 3))
comparation(ax, x_1, y_1, 'sigmoid(x)', x_2, y_2, 'tanh(x)')
```

```python
x = np.linspace(-10., 10., 100)
y = relu(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'ReLU(x)', x, y)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
d_relu_visualization(ax)
```

```python
x = np.linspace(-10., 10., 100)
y = leaky_relu(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'LReLU(x)', x, y)
```

```python
fig, ax = plt.subplots(figsize=(4, 3))
d_leaky_relu_visualization(ax)
```

```python
x = np.linspace(0., 1., 100)
y = softmax(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'Softmax(x)', x, y)
```

```python
x = np.linspace(0, 1, 100)
y = d_softmax(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'd_Softmax(x)', x, y)
```

```python
x = np.linspace(-3, 3, 100)
y = cross_entropy(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'CE(x)', x, y)
```

```python
x = np.linspace(-10, 10, 100)
y = e_power_x(x)

fig, ax = plt.subplots(figsize=(4, 3))
visualization(ax, 'f(x)', x, y)
```
